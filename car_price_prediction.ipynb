##**PRE PROCESSING**

###Lets calculate the age of the cars . With the present date .

```python
import datetime

date_time = datetime.datetime.now()
df['age'] = date_time.year - df['Year']
```

```python
df.head(1)
```

### So lets drop the year column from the data set coz we have new column that display the age of the car .

```python
df.drop(['Year'],axis = 1,inplace= True)
```

```python
df.head(1)
```

##**Let's check for the outlier**

```python
import seaborn as sns
sns.boxplot(df['Selling_Price'])
```

here these two points are so far from other data points .

```python
sorted(df['Selling_Price'],reverse=True)
```

```python
df = df[~(df['Selling_Price']>=33.0) & (df['Selling_Price']<= 35.5) ]
```

```python
df.shape
```


## Here the shape is reduced , hence we can concclude that the we removed the 2 outliers .

##**Encoding for the categorical data**

```python
data_types = df.dtypes
data_types
```

### As we can see that the car_name ,  fuel_type ,selling_type,Transmission are in the categorical values. Lets encode the Fuel_Type , first after that seller_type

```python
df['Fuel_Type'].unique()
```

```python
df['Fuel_Type']=df['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
df.Fuel_Type
```

###Lets encode the Seller_Type .

```python
df['Selling_type'].unique() # here we have two unique values .
```

```python
df['Selling_type'] =df['Selling_type'].map({'Dealer':0,'Individual':1})
```

```python
df.Selling_type
```

###Lets perform the same for the Transmission

```python
df['Transmission'].unique()
```

```python
df['Transmission'] = df['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
df.Transmission
```

```python
df.head()
```

##**Training and testing and spliting**

## For that lets drop 2 column names .1.car_name, selling_price .

```python
x= df.drop(['Car_Name','Selling_Price'],axis =1) # independent variables
y= df['Selling_Price'] #Target variables
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size =0.20,random_state =42)
```

##Lets find the best model for the prediction of the car price . for that we use grid search cv method

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
model1 = LinearRegression()
model1.fit(xtrain,ytrain)
```

```python
model2 = RandomForestRegressor()
model2.fit(xtrain,ytrain)
```

```python
model4 = XGBRegressor()
model4.fit(xtrain,ytrain)
```

```python
predict1 = model1.predict(xtest)
predict1
```

```python
predict2 = model2.predict(xtest)
predict2
```

```python
predict4 = model4.predict(xtest)
predict4
```

### Lets evaluate the performance of the algorithm used .

```python
from sklearn import metrics
score1= metrics.r2_score(ytest,predict1)
score1
```

```python
score2= metrics.r2_score(ytest,predict2)
score2
```

```python
score4= metrics.r2_score(ytest,predict4)
score4
```

```python
final = pd.DataFrame({'model':['LinearRegression','RandomForestRegressor','XGBRegresso'],'R2_Score':[score1,score2,score4]})
```

```python
final.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'final' is your DataFrame containing the results
sns.barplot(x='model', y='R2_Score', data=final)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.xlabel('Model')
plt.ylabel('R2 Score')
plt.title('R2 Score for Different Models')
plt.show()
```

### HENCE XGBRegresso is the best model for this problem .

```python
import matplotlib.pyplot as plt

# Assuming 'predict4' contains the predicted values and 'ytest' contains the actual values
plt.figure(figsize=(8, 6))
plt.scatter(ytest, predict4, alpha=0.5)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values (R2 Score: {:.2f})".format(score4))  # Provide the R2 score
plt.grid(True)
plt.show()
```
